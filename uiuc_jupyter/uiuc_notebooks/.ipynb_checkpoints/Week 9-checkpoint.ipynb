# Week 9: Additive Color

## RGB Additive Color

- Blue, Green and Red cones in the eye, each receptive to different wavelengths.


- RGB Color space is additive due to the three types of cones intermixing their signals.
- For instance yellow is an addition of red photons and green photons, but adding two cone responses
- Double the photons does not "look" twice as bright, though physically there are 2x # of photons
- Blue is not actually "darker" than red, it that is an illusion caused by our perception

- Each color may be thought of as an axis in the three dimensional color space
- Origin is black. Coordinates are values 0-1


### Equation for adding light.

- Essentially just add up the reflection from each suface (reflection operation), and find the pixel color, ignoring ambient light to do this
- Fuckin neat, this is like how all of astrophototgraphy works, duh
- Additive light from multiple sources, tinted by the color of lights within a scene allows for accurate reproductions of what a subject would look like in a given environment (sample the reference light color in order to get the correct value to filter light used for the final image



# Subtractive Color
- Cyan, Magenta, Yellow
- CMY Subtractive Color Space
- Surfaces are subtractive, they absorb photons. Others are reflected, and the remainder is the color that you see
- This is the basis of paint and ink colors (think your plotter)


- CMYK (black)
- CMY is an expensive color space to use for printing, so we use black in order to avoid having to use 1,1,1 in the CMY color space.
- Essentially get as close to the color as possible in black, then use colors to shift to actually coordinate




### RGB to CMY Transformations
- Perform an inversion of the matrix, and move the origin to (1,1,1) with a translation using a homogenous translation matrix

### CMY to CMYK
- Use a homogenous transformation matrix
- Identity matrix that subtracts the common amout of color, showing how much color to use off-black to get the desired color
