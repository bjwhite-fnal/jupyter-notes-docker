### Types of Jupyter notebook cells 
(1) **Markdown** cells (Cell->Cell Type->Markdown)
   - text, images 
   - cannot compile code you put in Markdown cells 
<br><br>

(2) **Code** cells (Cell->Cell Type->Code)
   - store and compile your code
   - any non-code text must be formatted as a comment - begin with "#"
<br><br>

### Running Jupyter notebook cells 
(1) **Entire notebook**
 - Cell->Run All
<br><br>

(2) **Single code cell**
 - Click on the cell and hit Shift-Enter 

```python
# Import the Python "modules" that we need.
# Many modules are standard, such that "import" is sufficient.
# Others require first building / compiling the software.
# We will use the Python "matplotlib" plotting module here.
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# We first read in a satellite image from a web URL.
# The function "imread" is used here.  The URL appears in red.
# The Python "variable" that holds the data here is called "satellite".
satellite = mpimg.imread('http://rfd.atmos.uiuc.edu/Atms517/twodisk.png')
```

```python
# Before we display the image we first determine its dimensions.
# We use the "shape" attributes of our "satellite" variable,
#   and print the X, Y dimensions of the image picture elements (pixels).
rows = satellite.shape[0]
columns = satellite.shape[1]
print('Satellite image dimensions are (%d x %d)' %
      ( columns, rows) )
```

```python
# Display the image, after specifying the size of the
# plot figure, and the dots-per-inch resolution we want to use.
# Afterwards, we annotate image with Matplotlib's title, xlabel, ylabel
plt.figure( figsize=(8,4.5),dpi=300 )
plt.imshow(satellite,interpolation='nearest')
plt.title('Current satellite images of Earth');
plt.xlabel('Columns increasing to right');
plt.ylabel('Rows increasing down');
```

```python
# Show the current date/time - Universal Coordinated Time (UTC)
from time import gmtime, strftime, localtime
strftime("Current day/time: %Y-%m-%d %H:%M:%S UTC", gmtime()) # UTC/GMT time
```
