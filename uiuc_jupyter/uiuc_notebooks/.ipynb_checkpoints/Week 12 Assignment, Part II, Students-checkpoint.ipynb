### !!!!! MINI LESSON ALERT  - ON RESAMPLING! !!!!!
(https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.resample.html)
- Put simply: resampling is used when you have data with a time dimension, and you want to get the data at either a higher or lower time frequency 
<br><br>
- Good examples: https://jakevdp.github.io/PythonDataScienceHandbook/03.11-working-with-time-series.html#Resampling-and-converting-frequencies
<br><br>
- df.resample('rule' (aka the frequency you want to resample by),optional arguments)
  - example: hourly data want to resample to daily frequency - df.resample('D'), where df is your Pandas DataFrame
  - frequency options: https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#dateoffset-objects
<br><br>
- resample changes the number of data points you have; smoothing does not.  Make sure you understand why!

#### (f.) To address the pattern on the shorter time-scale that you've noted in (e.), let's *resample* the data from daily data to monthly data, ultimately centered on the middle of the month 
**You will resample TWO different ways, as I explicitly describe below**
  - (1) resample CO2 measurements ('value') with monthly frequency- specifically *start of the month* 
      - **print your dataframe to prove to yourself that this worked - each monthly mean is at the first of each month**
<br><br>


```python
month_start = df.resample('MS').mean()
print(month_start.index)
```

  - (2) do the same as in (1), but instead anchor resampling to the *middle of the month*, as NOAA does for their plots
      - **print your dataframe to prove to yourself that this worked - each monthly mean is at 15th of each month**
      - see your 'Week 12 Assignment hints' file if you need these lines of code for (2) - a bit tricky and outside of the scope of your lessons

```python
month_mid = df.resample('MS', loffset=pd.Timedelta(14, 'd')).mean()
# What a user-unfriendly deprecation below. :/
```

#### (g.)  Now, let's plot your results of your resampling in (f.)
- (i.) create one plot, for 2016 through 2019, so we can really see smaller-scale details, containing the following:
    - the raw data - before any resampling
    - resampled data from (f.) anchored to start of month
    - resampled data from (f.) anchored to middle of month

- Don't forget: legend, labels, title, etc. on your plots

```python
fig, ax = plt.subplots(1, 1, figsize=(10,10))
year_s = 2106
year_e = 2019
df_range = df[(df.index.year>=2017) & (df.index.year<=2019)]
month_start_range = month_start[(month_start.index.year>=2017) & (month_start.index.year<=2019)]
month_mid_range = month_mid[(month_mid.index.year>=2017) & (month_mid.index.year<=2019)]

ax.scatter(df_range.index, df_range['value'], c='blue')
ax.scatter(month_start_range.index, month_start_range['value'], c='green')
ax.scatter(month_mid_range.index, month_mid_range['value'], c='red')
```

#### (h.)  Discuss, in at least a few sentences, conceptually what the distinction is between resampling and smoothing.

Resampling is the process by which the frequency by which the data has been measured is reaggregated and processed according to some form of time sequencing. Smoothing on the otherhand refers to the process why which noise is removed from a signal, enhancing the true nature of the underlying phenomenon. 

#### (i.)  Now we have our monthly averages via resampling - let's more easily be able to pull out longer term trends by SMOOTHING our CO2 data
- look closely at your data to choose the width of your window!
<br><br>
- you'll use your resampled data from method 2 (anchored at middle of month) for smoothing

```python
smoothed = month_mid.rolling("30D", center=True).mean()
```

#### (j.) Create a single plot (including labels, legend, etc) of CO2 data BEFORE resampling or smoothing vs. our monthly resampled data (anchored at middle of month)  vs. the smoothed, resampled data you just produced in (i.)

```python
fig, ax = plt.subplots(1, 1, figsize=(10,10))
ax.scatter(df.index, df['value'], c='black', alpha=1)
ax.scatter(smoothed.index, smoothed['value'], c='red', alpha=0.3)
```
