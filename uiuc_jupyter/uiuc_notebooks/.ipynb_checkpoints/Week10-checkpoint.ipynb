# Parametric Curves

## Cubic Curves
- Used for interpolation and surface moving
- Polynomials (3rd degree cubics) are well understood

- Cuvix are flexible enough to space curves where quadratics are limited to planar curves

## Wiggle Theorem
- y=x (ordinary line)
- Simplest form of interpolation
- Number of curves is 1-(degree of polynomial)

- 3rd degree has two wiggles

- Infinite wiggles (trigonometric functions where the taylor polynomial has an infinite number of terms)

## Lagrangian Interpolation
- Solve the cubic polynomial for points along the function

- Doesn't guarantee a shape for the resultant line, only guarantees that it passes through each point where we solve for 0
- Not actually a good way to go, due to the unbounded nature (no control over the shape of the graph)

# Hermite Curves
- Define a parametric funtion p(t)

- Decompose the function into two functions, one returning values of x the other returning values of y
- Each of these decompositions is its own interpolation between the values of x or y


- Now interpolate between both p0 and p1
- The formulas can be refactored to be based directly on p0, p1, and t


- So how do we do it for cubic curves?
- Two points with one line connecting them is easy for linear, harder for cubic interpolation
- So lets look at directions (i.e. derivative of the line at each point)
- Figures out the line that matches the start vector to the end vector each at their own given directions and rates
- Separate into coordinate functions, find a define a cubic function with no values yet

- Combine the x/y equations into a single equation that uses values of p

- Can be represented with linear algebra instead,

- Cubic basis functions can be refactored into cubic weights on the directional vectors


# Bezier Curves
- Easy to interpolate between p0 and p1 with parameter t which is going between p0 and p1 or between two coordinate functions (one for each axis)
- Decomposition of 2D interpolation into two 1D interpolations
- Basis functions are decompositions of the interpolation into formulas for each parameter
- Gets you weights for constant functions

## Bernstein Polynomials
- Weights of linear functions as opposed to constant functions

- Each parameter is weighted based on the sums

## Bernstein Interpolation
- These weights give you a bunch of functions that have differing weights between 0 and 1

- Multiply each of these basis curves by the x value


## Bernstein Polynomials

- Partition of unity. Anywhere we look between 0 and 1, the sum of the basis functions is 1
- Keeps us bounded between 0 and 1, with the area under all of the curves summing to 1 at all points


## Cubic Bezier Curves
- Bernstein basis applied to points

- NO DERIVITIVE REQUIRED
- The curve is specified by 4 CONTROL points including two ENDPOINTS
- Interpolate p0 to p3 controlled by p1 and p2
- 4 geomentric points that you have control over the location of
- "Control polygon"

- Control points are geometric, and work with affine transformations. Every point on the resultant curve is based off the control points
- Want to do affine transformation on curve? Just apply afine transformation to the control points

## de Casteljau Algorithm
- Split the line proportionally between p0 and p1 and find point p01
- Cascade this idea around the control polygon


- Do it again, connecting each of the new "middle" points, and continue on

- p0123 is on the curve at the exact value of t

- This proves that Bernstiens are substitutes for a bunch of linear interpolations with algebra
- Futher interpolation and cascading causes the subdivisions to converge to an approximation of the curve
- Can just plot the control polygon to get the curve
