# Texture Sampling

- Aliasing: the misidentification of a signal frequency, introducing distortion or error


### Minification Aliasing
- Aliases that happen due to the texture being sampled too sparsely

- in this case pulling from the direct pixel centers completely misses all the orange pixels, and only gets the blue
- Need a way to map a texture image color to an arbitrary number of texture image pixels
- Allows you to use the average of n pixels for one screen pixel
- This is done by using trilinear interpolation. Bilinear interpolation at two resolutions, and bilinearly interpolate between them to find the file trilinear interpolated color

#### MIP Mapping
- Progressively generate levels of average pixel color that we can index into based on the needed resolution
- Lance Williams




### Magnification Aliasing
- Happens when many image pixels sample a smaller number of texture pixels

#### Nearest Neighbor interpolation
- Each of the texture pixels is at the center of the pixels, but on the screen, you pick the color of the closest texture pixe corresponding to a image pixel
- Pretty fast as it uses only a 1 to 1 texture/image memory access


#### Bilinear interpolation
- Each pixel looks interpolates the 4 nearest pixels surrounding it
- Results in softer colors, but more accuracy
- Softer edges help to represent the fact that some of the pixels on the image lie between two texture image pixel centers
