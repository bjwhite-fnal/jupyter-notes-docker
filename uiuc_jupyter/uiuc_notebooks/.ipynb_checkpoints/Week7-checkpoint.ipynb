# Texture Mapping

- Need to fool the eye into seeing detail
- Do this by mapping images onto objects
- Generate coordinates (s, t) which are known as "texture coordinates (texcoords)" which map onto an image (x, y)

### Texture Coordinates

- There will be distortion due to having an image of a spherical earth indexed on a perfect spheroid

- Start with one triangle, and the grid of pixels

- Gives us the fragments associated with the triangle. Find (s, t) inside that triangle, which allows indexing into the image
