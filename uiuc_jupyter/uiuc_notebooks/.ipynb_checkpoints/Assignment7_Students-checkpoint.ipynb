### HELPFUL REMINDER FOR THIS ASSIGNMENT
- remember: dataframe['col name'] accesses a column
<br><br>

**EXAMPLE**
- let's say I have a dataframe consisting of columns of temperature, dew point, wind speed, and wind direction
<br><br>
- *dataframe[dataframe['temperature']>50]* -> displays any rows (spanning over all columns) in dataframe whose temperature exceeds 50
  - there are other ways to do this exact same thing
<br><br>
- what if I only want the dewpoint column of the results?
   - *dataframe[dataframe['temperature']>50]['dew point']* 
      - this is one of many ways of doing this
   - alternatively: *dataframe[dataframe['temperature']>50].loc[:,'dewpoint']* - all rows and just the dewpoint column

#### (2)  What month and day did the strongest earthquake occur, and in what country?
- You do NOT have to parse the date string, or the place string... for now, you'll just work with the full strings 
- FYI, how to (manually) read date string: year-month-dayTtime

```python
max_mag_idx = tdata['mag'].idxmax()
max_time = tdata.loc[max_mag_idx]['time']
max_country = tdata.loc[max_mag_idx]['place']
from dateutil import parser
date = parser.parse(max_time)
print(f'The bigliest earthquake occured on {date.month}/{date.day} {max_country}')
```

#### (3) How many earthquakes occur with magnitude greater than 6?

```python
num_over_6 = len(tdata.query('mag > 6'))
print(f'There were {num_over_6} earthquakes with magnitudes greater than 6')
```

#### (4) What is the avearge depth of earthquakes with magnitude greater than 6?

```python
avg_depth = tdata.query('mag > 6')['depth'].mean()
print(f'There average depth of earthquakes greater than magnitude 6 was: {avg_depth:.2f}')
```

#### (5) How many earthquakes occur with magnitude greater than 6 and depth less than 30 km?

```python
count = len(tdata.query('mag > 6 and depth < 30'))
print(f'There were {count} earthquakes greater than magnitude 6 with a depth less than 30km')
```

- **CHALLENGED, UNGRADED**:  color-code markers by magnitude and/or size in a way that really offsets the more significant earthquakes from the many, many weaker earthquakes (we'll colorcode by two groups of magnitudes in the next question, but I'm asking for a range of color over all magnitudes, if that makes sense)

```python
lon = tdata['longitude']
lat = tdata['latitude']
X, Y = np.meshgrid(lon, lat)
proj = ccrs.PlateCarree()
fig, ax = plt.subplots(figsize=(15,15), subplot_kw=dict(projection=proj))
im = ax.scatter(lon, lat, c=tdata['mag'], cmap='coolwarm')
fig.colorbar(im, ax=ax,fraction=0.046, pad=0.04)
gl = ax.gridlines(draw_labels=True)
gl.top_labels = False
gl.right_labels = False
ax.coastlines()
ax.set_title('')
ax.set_title('Global Earthquake Magnitudes')
```

#### (8) Make the same map as in (6), but instead, plot earthquakes with magnitude greater than 6 a different color than earthquakes with magnitude less than or equal to 6, and a different size (stronger earthquakes = larger markers)

```python
over_6 = tdata.query('mag > 6')
under_6 = tdata.query('mag <= 6')
olon = over_6['longitude']
olat = over_6['latitude']
ulon = under_6['longitude']
ulat = under_6['latitude']
X, Y = np.meshgrid(lon, lat)
proj = ccrs.PlateCarree()
fig, ax = plt.subplots(figsize=(15,15), subplot_kw=dict(projection=proj))
mags = tdata['mag'].tolist()
sizes = []
colors = []
for mag in mags:
    if mag > 6:
        sizes.append(mag**3)
        colors.append('r')
    else:
        sizes.append(mag)
        colors.append('b')
im = ax.scatter(lon, lat, c=colors, s=sizes)
gl = ax.gridlines(draw_labels=True)
gl.top_labels = False
gl.right_labels = False
ax.coastlines()
ax.set_title('Global Earthquake Magnitudes Over/Under 6.0')
```
