### !!!!! MINI LESSON ALERT  - ON RESAMPLING! !!!!!
(https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.resample.html)
- Put simply: resampling is used when you have data with a time dimension, and you want to get the data at either a higher or lower time frequency 
<br><br>
- Good examples: https://jakevdp.github.io/PythonDataScienceHandbook/03.11-working-with-time-series.html#Resampling-and-converting-frequencies
<br><br>
- df.resample('rule' (aka the frequency you want to resample by),optional arguments)
  - example: hourly data want to resample to daily frequency - df.resample('D'), where df is your Pandas DataFrame
  - frequency options: https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#dateoffset-objects
<br><br>
- resample changes the number of data points you have; smoothing does not.  Make sure you understand why!
