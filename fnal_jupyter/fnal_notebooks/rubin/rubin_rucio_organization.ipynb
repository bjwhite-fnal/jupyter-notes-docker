# Scopes

### Scope: `raw`

- Represents raw observation data from any instrument, whether on-sky, calibration, engineering, or on a test stand.
Includes precursor on-sky data, if desired.
In any case, includes data from multiple instruments, such as the three Rubin Observatory ones.

- Manipulations of this scope need to be highly restricted.

### Scope: `simulation`

- Manipulations of this scope are less restricted than for raw data.

### Scope: `products`

- Manipulations of this scope are inherently multi-site and open to a wider group.

- If we want to avoid container-level permissions and only have scope-level permissions, then DPn/DRn needs to be promoted to this level.


# Containers


###  Scopes `raw` and `simulation`

**First level:** instrument or test stand name or "LFA" or other non-Butler label

- This container level separates out the data by source.
It could correspond to permissions for certain accounts to manipulate data from that source, although it's unlikely we will need that level of granularity.
- This can only be closed if a test stand or non-Butler data type is eliminated.

**Second level:** LFA source for LFA containers

- This further separates out non-Butler-ingested LFA sources.
- This can only be closed if the source no longer exists.

### Scope `products`


**First level:** DPn or DRn

- This container level separates out the data by reprocessing.
In can correspond to permissions for certain accounts to manipulate data for that reprocessing (and not others), using a custom permissions policy.

- These can be closed when the given processing is released.
Any patches thereafter could be DR1.1, etc.

**Second level:** Butler run collection

- This container level separates out the data by run in order to partition them to a reasonable size.
- These can be closed when the given run is complete.


# Datasets

### Scopes `raw` and `simulation`

**Name:** observation date

- These datasets can be monotonic.
- If they cannot hold millions of files, observation hour may need to be added.

- These can be closed when the given date/hour is complete, including any catch-up archiving.

### Scope `products`

**Name:** dataset type

- If they cannot hold millions of files, additional Butler dimensions such as band, visit, tract, patch, etc. may need to be added.

- These can be closed when all products of the given type have been generated, according to the relevant quantum graph/Campaign Management.
